# Que. Create pipeline, remove stop words, word frequency analysis and word clouds

## Pre-Processing Pipeline

```python
# Importing Regex Package
import regex as re
# Defining a Function to Pull Words
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
# Importing nltk Package
import nltk
```

```python
# Downloading all List of Stop Words
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
# Creating a Pipeline
pipeline = [str.lower, tokenize]

# Defining a Function to Run the Pipeline
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
# Running the Pipeline
df['tokens'] = df['Reviews'].apply(prepare, pipeline=pipeline)
```

```python
# Creating a New Column for the Number of Tokens
df['num_tokens'] = df['tokens'].map(len)
df.head()
```

## Word Frequency Analysis

```python
# Importing a function Counter
from collections import Counter

counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
# Creating a New Data Frame for Frequency Analysis 
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

```python
# Sorting the freq_df based on Frequency
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
# Printing the Statistical Data of Frequency
freq_df.describe()
```

### We are using a function cleanTxt that takes a text input and performs several cleaning operations on it. The cleaning operations include removing mentions (text starting with '@'), removing hashtags ('#'), removing retweets (text starting with 'RT'), removing hyperlinks, and removing colons

```python
# Clean the text

# Create a function to clean the tweets
def cleanTxt(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) #Remove @mentions replace with blank
    text = re.sub(r'#', '', text) #Remove the '#' symbol, replace with blank
    text = re.sub(r'RT[\s]+', '', text) #Removing RT, replace with blank
    text = re.sub(r'https?:\/\/\S+', '', text) #Remove the hyperlinks
    text = re.sub(r':', '', text) # Remove :
    
    return text

# Cleaning the text
df["Reviews"]= df["Reviews"].apply(cleanTxt)

# Show the clean text
df.head()
```

### We are creating a function named remove_emoji that uses a regular expression to match and remove various Unicode ranges representing emojis and symbols

```python
# Next we have to remove emoji & Unicode from the reddit data.
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

# Cleaning the text
df["Reviews"]= df["Reviews"].apply(remove_emoji)

# Show the clean text
df.head()
```

### Installing TextBlob, which is a Python library for processing textual data

```python
from textblob import TextBlob
from wordcloud import WordCloud, STOPWORDS
```

```python
# Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Create a function to get Polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

# Now we create a new column for what we just did and add it to the Tweet_df dataframe
df['Subjectivity'] = df['Reviews'].apply(getSubjectivity)
df['Polarity'] = df['Reviews'].apply(getPolarity)

df.head()
```

```python
# Group the range of Polarity to different categories
def getInsight(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
    
df["Insight"] = df["Polarity"].apply(getInsight)

df.head(50)
```

### Importing necessary libraries and utilizing the NLTK library to access the opinion lexicon. The code checks the total number of words in the opinion lexicon and prints examples of positive and negative words

```python
# Import necessary Libraries
from sklearn import preprocessing
import nltk
nltk.download('opinion_lexicon')
from nltk.corpus import opinion_lexicon
from nltk.tokenize import word_tokenize

print('Total number of words in opinion lexicon', len(opinion_lexicon.words()))
print('Examples of positive words in opinion lexicon',
      opinion_lexicon.positive()[:10])
print('Examples of negative words in opinion lexicon',
      opinion_lexicon.negative()[:10])
```

```python
# Let's create a dictionary which we can use for scoring our review text
nltk.download('punkt')
df.rename(columns={"reviewText": "text"}, inplace=True)
pos_score = 1
neg_score = -1
word_dict = {}
 
# Adding the positive words to the dictionary
for word in opinion_lexicon.positive():
        word_dict[word] = pos_score
      
# Adding the negative words to the dictionary
for word in opinion_lexicon.negative():
        word_dict[word] = neg_score
```

### The function returns the calculated sentiment score for the input text

```python
def bing_liu_score(text):
    sentiment_score = 0
    bag_of_words = word_tokenize(text.lower())
    for word in bag_of_words:
        if word in word_dict:
            sentiment_score += word_dict[word]
    return sentiment_score  
```

```python
df['Bing_Liu_Score'] = df['Reviews'].apply(bing_liu_score)
```
